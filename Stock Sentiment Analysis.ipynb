Stock Sentiment Analysis adalah sebuah eksprimen yang saya lakukan dengan mengikuti tutorial [Krish Naik](https://www.youtube.com/watch?v=h-LGjJ_oANs&list=PLZoTAELRMXVMdJ5sqbCK2LiM0HhQVWNzm&index=12).<br><br>
* Dataset yang digunakan adalah dataset yang didapat dari gabungan antara berita berita diseluruh dunia yang terkait dengan *Stock Price* (Harga Saham) dan juga dataset Stock Price yang didapat dari <b>Kaggle</b>
* Data terdiri dari rentang waktu 2008 s/d 2016 dan data dari rentang tahun 2000 sampai 2008 didapat dari hasil *scrapping* pada *Yahoo Finance*
* Label didapat dari [*Dow Jones Industrial Average Index*](https://www.google.com/search?q=dow+jones+average+stoc+index&oq=dow+jones+average+stock+index&aqs=chrome..69i57j0i22i30j0i8i13i30l2j69i64.1950j0j7&sourceid=chrome&ie=UTF-8)
* Label 1 menyatakan harga saham naik
* Label 0 menyatakan harga saham tetap atau turun

```python
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/krishnaik06/Stock-Sentiment-Analysis/master/Data.csv'
df = pd.read_csv(path, encoding='ISO-8859-1')
df.head()
```

#### Memisahkan Train dan Test data

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

### Preprocessing

#### Cleaning the text

```python
# menghilangkan punctuation
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]"," ", regex=True, inplace=True)

# mengganti nama kolom
list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head()
```

```python
# mengubah setiap kata pada teks menjadi lower case
for idx in new_index:
    data[idx] = data[idx].str.lower()
data.head(1)
```

```python
# menjadikan setiap teks dalam satu baris ke dalam satu paragraf
headlines = []
for row in range(0, len(data.index)):
    headlines.append(''.join(str(x) for x in data.iloc[row, 0:25]))
    
```

```python
headlines[0]
```

#### Tokenizing dan Implementasi model *Bag of Words*

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(2,2))
traindata = cv.fit_transform(headlines)
```

### Proses Klasifikasi

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200,criterion='entropy')
rf.fit(traindata, train['Label'])
```

### Proses Prediksi

```python
# menggabungkan kata data setiap baris data test ke dalam satu paragraf
test_transform = []
for row in range(0, len(test.index)):
    test_transform.append(''.join(str(x) for x in test.iloc[row, 0:25]))
    
# implementasi Bag of Words pada data test
testdata = cv.transform(test_transform)
predict = rf.predict(testdata)
```

### Evaluasi Model

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predict)
print('Confusion Matrix : \n', matrix, '\n')

acc=accuracy_score(test['Label'],predict)
print('Accuracy = ', acc, '\n')
report=classification_report(test['Label'],predict)
print(report)
```
